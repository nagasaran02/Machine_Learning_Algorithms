# Import Required Libraries

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np
```

# Load the Iris dataset

```python
iris = load_iris()
# The input features (sepal length, sepal width, petal length, petal width)
X = iris.data  
```

# Apply DBSCAN to the Iris dataset


```python
dbscan_iris = DBSCAN(eps=0.5, min_samples=5)
dbscan_iris_labels = dbscan_iris.fit_predict(X)
```

# Visualize the clusters on the Iris dataset 

```python
plt.figure(figsize=(8, 6))
unique_labels = np.unique(dbscan_iris_labels)
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
```

# Plot each cluster with different color

```python
for i, label in enumerate(unique_labels):
    if label == -1:  # Noise points
        color = 'k'
        marker = 'x'
    else:
        color = colors[i % len(colors)]
        marker = 'o'
    plt.scatter(X[dbscan_iris_labels == label, 2],
                X[dbscan_iris_labels == label, 3],
                c=color, marker=marker, 
                label=f'Cluster {label}' if label != -1 else 'Noise',
                edgecolor='k')

# Add plot titles and labels
plt.title('DBSCAN Clustering on Iris Dataset (Petal Length and Width)')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Importing necessary libraries
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import numpy as np

# Generate a dataset (2D) with moons pattern to show DBSCAN's power in finding clusters of arbitrary shape
X, y = make_moons(n_samples=300, noise=0.05, random_state=42)

# Apply DBSCAN with epsilon (eps) and minimum points (min_samples)
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Plot the results
plt.figure(figsize=(8, 6))
unique_labels = np.unique(dbscan_labels)
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']

# Plot each cluster with different color
for i, label in enumerate(unique_labels):
    if label == -1:  # Noise points
        color = 'k'
        marker = 'x'
    else:
        color = colors[i % len(colors)]
        marker = 'o'
    plt.scatter(X[dbscan_labels == label, 0], X[dbscan_labels == label, 1], c=color, marker=marker, label=f'Cluster {label}' if label != -1 else 'Noise', edgecolor='k')

# Add plot titles and labels
plt.title('DBSCAN Clustering on Moons Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
