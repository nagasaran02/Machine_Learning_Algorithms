# K Means Clustering

# Importing necessary libraries

```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the iris dataset

```python
iris = load_iris()
# The input features (sepal length, sepal width, petal length, petal width)
X = iris.data 
# The true labels (for reference)
y = iris.target 
```

# Initialize KMeans with 3 clusters (since we know there are 3 species of iris)

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

# Predict cluster labels

```python
cluster_labels = kmeans.predict(X)
```

# Convert to DataFrame for visualization

```python
iris_df = pd.DataFrame(X, columns=iris.feature_names)
iris_df['Cluster'] = cluster_labels
iris_df['True Label'] = y
```

# Visualize the clusters (using only two features for easy visualization: petal length and petal width)


```python
plt.figure(figsize=(8, 6))
plt.scatter(iris_df.iloc[:, 2], iris_df.iloc[:, 3], 
            c=cluster_labels, 
            cmap='viridis', 
            marker='o', 
            edgecolor='k')
plt.title('K-Means Clustering of Iris Dataset')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Step 1: Generate sample data (for demonstration purposes)
# Create a dataset with 4 clusters
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Step 3: Plot the clustered data
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Step 4: Plot the centroids of the clusters
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')

plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
